```python
import pandas as pd
import numpy as np
import csv
import sklearn
import math
import scipy

from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn import preprocessing as pp
```

```python
pd.set_option('display.max_rows', None)
def showcsv(src):
    df = pd.read_csv(src)
    df.head()
    print(df)
    
def colsum(src):
    df = pd.read_csv(src)
    df = df.drop('timestamp', axis=1)
    colsum = df.sum(axis=0)
    print(colsum)
```

```python
def transformcsv(src,val,target,target2,newhdr):
    df = pd.read_csv(src)
    df['hourly_traffic_count']=df['hourly_traffic_count']*val

    df.to_csv('DataSet/dataset/New Folder'+target)
    
    with open('DataSet/dataset/New Folder'+target, newline='') as infile, open('new/split/'+target2, 'w', newline='') as outfile:
        r = csv.reader(infile)
        w = csv.writer(outfile)

        next(r, None) 
        w.writerow(['timestamp',newhdr])

        for row in r:
            w.writerow(row)
            
    print(pd.read_csv('new/split/'+target2))
    df = df.drop('timestamp', axis=1)
    colsum = df.sum(axis=0)
    print(colsum)
```

```python
transformcsv('DataSet/dataset/train26804.csv',0.4,'mc1.csv','MC.csv','Qmc')
```

```python
transformcsv('DataSet/dataset/train16020.csv',1.3,'hv1.csv','HV.csv','Qhv')
```

```python
transformcsv('DataSet/dataset/train26803.csv',1.0,'lv1.csv','LV.csv','Qlv')
```

```python
df1 = pd.read_csv('new/split/MC.csv')
df2 = pd.read_csv('new/split/HV.csv')
df3 = pd.read_csv('new/split/LV.csv')
df4 = pd.merge(df1, df2)
dfm = pd.merge(df4, df3)

print(dfm)
```

```python
df=list(dfm)
df.remove("timestamp")
dfm["Qmv"]=dfm[df].sum(axis=1)

print(dfm)
dfm.to_csv('new/merged.csv')
```
